```python
# Bu dersimizde örnek olarak kullanacagimiz veri seti yapay ögrenme alaninin en popüler veri setlerinden “Iris” veri seti.
# Iris veri seti 3 Iris bitki türüne (Iris setosa, Iris virginica ve Iris versicolor) ait, 
# her bir türden 50 örnek olmak üzere toplam 150 örnek sayisina sahip bir veri setidir. 
# Her bir örnek için 4 özellik tanimlanmistir: taç yaprak uzunlugu, taç yaprak genisligi, çanak yaprak genisligi,
# çanak yaprak uzunluğu('sepal length','sepal width','petal length','petal width'). 
# Veri setimizde, her bir bitki örnegi ayri bir gözlemi (örnegi) ifade ederken; 
# bitki tür ismi bagimli(dependent) degisken, bitkilerin ölçülen 4 temel özelligi ise
# bagimsiz(independent) degiskenleri ifade eder.
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
url="pca_iris.data"
# datasetimizi pandas datafrsme içine yüklüyoruz.
df=pd.read_csv(url,names=["sepal length","sepal width","petal length","petal width","target"])
df
```

```python
features=["sepal length","sepal width","petal length","petal width"]
```

```python
# feature'ları x olarak ayıralım
```

```python
x=df[features]
```

```python
# target'ı y olarak ayıralım
```

```python
y=df[["target"]]
```

```python
# degerleri scaler etmemiz gerekiyor.çünkü her bir featur farklı boyutlarda bu amaçla standart scalaer kullanarak tüm 
# verileri mean=0 and variance=1 olarak değiştiriyoruz.
```

```python
# standardizing the features

x= StandardScaler().fit_transform(x)
x
```

```python
# Biz PCA yaparak bunları 2 boyuta indirgeyeceğiz ancak şunu belirtmeliyim ki PCA indirgeme işlemi sonucunda 
# elde edeceğimiz 2 boyutun herhangi bir anlam ifade etmeyen başlıklara sahip olacak.. 
# Yani 4 feature'dan 2 tanesini basit bir şekilde atmak değil yaptığımız
```

```python
pca=PCA(n_components=2)
principalComponents=pca.fit_transform(x)
principalDf=pd.DataFrame(data = principalComponents, columns=["principal component 1","principal component 2"])
                         
```

```python
principalDf
```

```python
final_dataframe=pd.concat([principalDf,df[["target"]]],axis=1)
```

```python
final_dataframe.head()
```

```python
dfsetosa=final_dataframe[df.target=="Iris-setosa"]
dfvirginica=final_dataframe[df.target=="Iris-virginica"]
dfversicolor=final_dataframe[df.target=="Iris-versicolor"]
plt.xlabel("principal component1")
plt.ylabel("principal component2")

plt.scatter(dfsetosa["principal component 1"],dfsetosa["principal component 2"],color="green")
plt.scatter(dfvirginica["principal component 1"],dfvirginica["principal component 2"],color="red")
plt.scatter(dfversicolor["principal component 1"],dfversicolor["principal component 2"],color="blue")
```

```python
# Daha profesyonel bir çizim yapalım for  ile
```

```python
targets=["Iris-setosa","Iris-versicolor","Iris-virginica"]
colors=["g","b","r"]

plt.xlabel("princibal component 1")
plt.ylabel("princibal component 2")

for target, col in zip (targets,colors):
    dftemp= final_dataframe[df.target==target]
    plt.scatter(dftemp["principal component 1"], dftemp["principal component 2"],color=col)
```
